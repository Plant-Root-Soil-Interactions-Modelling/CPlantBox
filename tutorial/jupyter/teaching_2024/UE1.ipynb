```python
import numpy as np
import matplotlib.pyplot as plt
theta_s=0.35;
theta_r=0.01;
theta=np.linspace(theta_r,theta_s,50);
Se=(theta-theta_r)/(theta_s-theta_r);
plt.plot(theta,Se,'k-');
plt.xlabel('Wassergehalt (-)',fontsize=20);
plt.ylabel('Relativer Wassergehalt (-)',fontsize=20);
plt.show()
```

# Aufgabe 3

```python
import numpy as np
import matplotlib.pyplot as plt
#Fitting Parameter
ha=-10
lam=0.5
theta_r=0.01
theta_s=0.4
#Variables
h=-np.logspace(-2,5,1000);
theta = np.ones(len(h))*theta_s;
for i in range(len(h)):
    if h[i]<ha:
        theta[i] = ((ha/h[i])**lam)*(theta_s-theta_r)+theta_r
#Plot
plt.semilogx(abs(h),theta,'k-')
plt.xlabel('Saugspannung (cm)',fontsize=10)
plt.ylabel('Wassergehalt (-)',fontsize=10)
plt.title('Brooks and Corey Kurve',fontsize=15)
plt.xscale('log')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Fitting Parameter
theta_s=0.4;
theta_r=0.01;
n=1.5;
alpha=0.8;
ha=-10;
lam=0.5;
# independent variable h
h=-np.logspace(-2,5,100);
# dependent variable theta
thetabc=np.ones((100))*theta_s;
thetabc[h<ha]=(ha/h[h<ha])**lam * (theta_s - theta_r) + theta_r;
#for i in range(len(thetabc)):
#    if h[i]<ha:
#        thetabc[i]=(ha/h[i])**lam * (theta_s - theta_r) + theta_r;
thetavg=(theta_s-theta_r)/(1+(-alpha*h)**n)**(1-1/n)+theta_r;
thetac=np.ones((100))*theta_s;
b=1.6
thetac[h<ha] = theta_s*(h[h<ha]/ha)**(-1/b)
# Plot
plt.semilogx(abs(h),thetabc,'k-',label="Brooks & Corey");
plt.semilogx(abs(h),thetavg,'k--',label="Van Genuchten");
plt.semilogx(abs(h),thetac,'g--',label="Campbell");
plt.xlabel('Saugspannung (cm)',fontsize=20);
plt.ylabel('Wassergehalt (-)',fontsize=20);
plt.legend()
plt.show()
```

# Aufgabe 4

```python
import numpy as np
import matplotlib.pyplot as plt
Se=np.linspace(0.001,1,100);
# Sand
n=1.46;
m=1-1/n;
l=0.5;
Ks=146.6; # cm h-1
K_sand=Ks*Se**l*(1-(1-Se**(1/m))**m)**2;
# Lehm
n=1.25;
m=1-1/n;
l=0.5;
Ks=31.68; # cm h-1
K_lehm=Ks*Se**l*(1-(1-Se**(1/m))**m)**2;
# Ton
n=1.08;
m=1-1/n;
l=0.5;
Ks=1.44; # cm h-1
K_ton=Ks*Se**l*(1-(1-Se**(1/m))**m)**2;
#Plotplt.semilogy(Se,K_sand,'k-',label="Sand");
plt.semilogy(Se,K_lehm,'k--',label="Lehm");
plt.semilogy(Se,K_ton,'k-.',label="Ton");
plt.xlabel(r'$\mathrm{Relative S\ddot{a}ttigung (-)}$',fontsize=20);
plt.ylabel(r'$\mathrm{Hydraulische Leitf\ddot{a}higkeit (cm h^{-1})}$',fontsize=20);
plt.legend(loc=4)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# ModelParameter
r=1.5; #cm/d
k=30; #cm
#Variables
t=np.linspace(0,50,100);
l=k*(1-np.exp(-r/k*t));
l_linear = r*t;
l_linear[l_linear>k]=k;
#Plot
plt.figure;
plt.plot(t,l,'k-',label="Negative expoential growth");
plt.plot(t,l_linear,'k--',label="Linear")
plt.xlabel('Zeit t (days)',fontsize=20);
plt.ylabel('Wurzellänge (cm)',fontsize=20);
plt.legend(loc=2)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
Cl=np.linspace(0.0,0.1,100);
# ModelParameter
Imax=17.3e-8
Km=1.6e-3
I1 = Imax*Cl / (Km + Cl);
Imax=3.7e-8
Km=1.0e-3
I2 = Imax*Cl / (Km + Cl);
# ModelParameter
Imax=17.3e-8
Km=1.6e-3*10
I3 = Imax*Cl / (Km + Cl);
Imax=3.7e-8
Km=1.0e-3*10
I4 = Imax*Cl / (Km + Cl);
hfont = {'fontname':'Arial'}
#Plot
fig=plt.figure(figsize=(15,8));
ax = fig.add_subplot(1,2,1)
plt.plot(Cl,I1,'DarkOrange',linewidth=2,label="P-Mangel");
plt.plot(Cl,I2,'k--',linewidth=2,label="P ausreichend");
plt.xlabel('Cl ($\mu$mol cm$^{-3}$)',fontsize=20);
plt.ylabel('I ($\mu$mol cm$^{-1}$ s$^{-1}$)',fontsize=20);
plt.title('Michaelis Menten Kinetik',fontsize=24);
plt.legend(loc=2)
ax = fig.add_subplot(1,2,2)
plt.plot(Cl,I3,'DarkOrange',linewidth=2,label="P-Mangel,Km*10");
plt.plot(Cl,I4,'k--',linewidth=2,label="P ausreichend,Km*10");
plt.xlabel('Cl ($\mu$mol cm$^{-3}$)',fontsize=20,**hfont);
plt.ylabel('I ($\mu$mol cm$^{-1}$ s$^{-1}$)',fontsize=20);
plt.title('Parameter-Variation',fontsize=24);
plt.legend(loc=2)
plt.xticks(size=20,**hfont)
plt.ylim((0,1.6e-7))
plt.show()
```
