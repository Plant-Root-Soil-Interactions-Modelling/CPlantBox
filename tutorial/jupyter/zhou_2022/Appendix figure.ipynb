```python
import plotly.express as px
import pandas as pd
# df = pd.read_csv("small_2020.xml_31_0.1_0.1_0.1.csv")
# df= pd.read_csv("small_2020_abs.xml_31_0.1_0.1_0.1_30.csv")
# df = pd.read_csv("case_1.1.1_0.1cm.xml_11_0.8_0.8_11_10.csv")
def connect_fig(name, time, attri, title, unit):
    df = pd.read_csv(name)
    # df = pd.read_csv("case_1_lateral_radius.xml_30_0.8_0.8_11_29.csv"

    df_1 = df[df["time"] == time]
    c1 = df_1["connection_1"].tolist()
    c2 = df_1["connection_2"][:-1].astype(int).tolist()
    radius_n = df_1['radius'].tolist()

    cx = df_1["coor_x"].tolist()
    cy = df_1["coor_y"].tolist()
    cz= df_1["coor_z"].tolist()
    rst= df_1["r_st"].tolist()
    rxyl= df_1["r_xyl"].tolist()
    cc= df_1["connect_no"].tolist()
    jwxyl= df_1["JW_Xyl"].tolist()
    jwst= df_1["JW_ST"].tolist()
    jsst= df_1["JS_ST"].tolist()
    radius =list(map(lambda x:radius_n[x-1],  c2 ))
    # cxyl1 = map(lambda x: x/10, cxyl)

    con_x = []
    con_y = []
    con_z = []
    ccc = []
    for i in range(0, len(c1)-1):
        con_x.append( (cx[int(c1[i]-1)] + cx[int(c2[i]-1)] )/2    )
        con_y.append( (cy[int(c1[i]-1)] + cy[int(c2[i]-1)] )/2    )
        con_z.append( (cz[int(c1[i]-1)] + cz[int(c2[i]-1)] )/2    )
        ccc.append(cc[int(c1[i]-1)])
    df2 = pd.DataFrame(list(zip(con_x, con_y,con_z, rst, rxyl,ccc, radius,jwxyl,jsst)), 
                   columns =['con_x', 'con_y','con_z', 'rst', 'xylem_axial_resistance','ccc','radius', 'axial_water','phloem carbon flow']) 

    fig = px.scatter_3d(df2, x="con_x", y="con_y", z = "con_z" ,
               size=radius, color=attri, color_continuous_scale=px.colors.sequential.Plasma_r,
               log_x=False)

    fig.update_traces(marker=dict(
                                  line=dict(width=0,
                                            color='DarkSlateGrey')),
                      selector=dict(mode='markers'))
    
    
    
    scence = dict(xaxis_title='X AXIS ',
                            yaxis_title='Y AXIS ',
                            zaxis_title='Z AXIS ',
                        xaxis = dict(
                             backgroundcolor="rgb(250, 250, 250)",
                             gridcolor="lightgrey",
                             showbackground=True,
                             zerolinecolor="black",),
                        yaxis = dict(
                            backgroundcolor="rgb(250, 250,250)",
                            gridcolor="lightgrey",
                            showbackground=True,
                            zerolinecolor="black"),
                        zaxis = dict(
                            backgroundcolor="rgb(250, 250,250)",
                            gridcolor="lightgrey",
                            showbackground=True,
                            zerolinecolor="black",),)


    fig.update_layout(coloraxis_colorbar=dict(
        title=dict(
            text='{}'.format(unit),
            font=dict(
                family="Arial",
                size=28,
                color='#000000'
            )),
        thicknessmode="pixels", thickness=16,
        lenmode="pixels", len=350,
        tickfont = {"size": 18}


    ))
    camera = dict(
        up=dict(x=0, y=0, z=1),
        center=dict(x=0, y=0, z=0),
        eye=dict(x=0, y=-2, z=0)
    )
    scence = dict(
        xaxis_title='',
                            yaxis_title='',
                            zaxis_title='',
                        xaxis = dict(
                             backgroundcolor="rgb(250, 250, 250)",
                             gridcolor="lightgrey",
                             showbackground=True,
                             zerolinecolor="black",
                             range=[-0.11,0.11],
                        showticklabels=False),
                        yaxis = dict(
                            backgroundcolor="rgb(250, 250,250)",
                            gridcolor="lightgrey",
                            showbackground=True,
                            zerolinecolor="black",
                            range=[-0.11,0.11],
                        showticklabels=False),
                        zaxis = dict(
                            backgroundcolor="rgb(250, 250,250)",
                            gridcolor="lightgrey",
                            showbackground=True,
                            zerolinecolor="black",
                            range=[-0.24,0.13],
                        showticklabels=False),)
    fig.update_layout(autosize=False,  scene = scence, scene_camera= camera, width =600,
                            height = 800,
                            margin=dict(r=0, b=0, l=0, t=60), title=dict(
            text='<b>{}</b> {}'.format(time, title),
            x=0.5,
            y=0.95,
            font=dict(
                family="Arial",
                size=36,
                color='#000000'
            )))
    fig.update_layout(scene_aspectmode='manual', scene_aspectratio=dict(x=1, y=1, z=2))
    fig.show()
```
