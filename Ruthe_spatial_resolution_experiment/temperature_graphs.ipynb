# Create graphs from the soil and air temperatures of the Ruthe field site

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.dates import DateFormatter, MonthLocator
from pathlib import Path

BASE_DIR = Path.cwd()
```

## Growth period November 1994 to August 1995

### Soil temperature

```python
# Soil temperature for 1995 in the depths 10 to 50cm (data before 24.06.95 is simulated)
df_soil = pd.read_pickle(f"{BASE_DIR}/data/data_processed/bodentemp_95.pkl")

df_soil.index = pd.to_datetime(df_soil.index)
df_soil.columns = df_soil.columns.astype(int)

depths = [10, 20, 30, 40, 50]

fig, ax = plt.subplots(figsize=(12, 6))

for depth in depths:
    ax.plot(df_soil.index, df_soil[depth],
            linewidth=2,
            alpha=0.85,
            label=f"−{depth} cm")

ax.grid(True, linestyle="--", alpha=0.4)

ax.xaxis.set_major_formatter(DateFormatter('%b %Y'))
plt.xticks(rotation=45, fontsize=16)
plt.yticks(fontsize=16)

#plt.xlabel("Month", fontsize=18)
plt.ylabel("Soil Temperature (°C)", fontsize=18)
#plt.title("Soil Temperature at Different Depths (Nov 1994 – Aug 1995)", fontsize=20)
plt.legend(title="Depth", fontsize=18, title_fontsize=18, loc="lower right")
plt.tight_layout()
plt.show()
```

### Air temperature

```python
# Air temperature for the growth period November 1994 to August 1995
df_1994 = pd.read_csv(f"{BASE_DIR}/data/data_processed/climate_data_94.csv")
df_1995 = pd.read_csv(f"{BASE_DIR}/data/data_processed/climate_data_95.csv")

df_1994['Date'] = pd.to_datetime('1994-01-01') + pd.to_timedelta(df_1994['day_of_year'] - 1, unit='D')
df_1995['Date'] = pd.to_datetime('1995-01-01') + pd.to_timedelta(df_1995['day_of_year'] - 1, unit='D')

df_1994 = df_1994[df_1994['Date'] >= '1994-11-01']
df_1995 = df_1995[df_1995['Date'] <= '1995-08-31']

df = pd.concat([df_1994, df_1995], ignore_index=True)

df = df.set_index('Date')

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(df.index, df['air_temp_average'])

ax.xaxis.set_major_formatter(DateFormatter('%b'))
plt.xticks(rotation=45)

plt.xlabel("Month")
plt.ylabel("Average Air Temperature (°C)")
plt.title("Average Air Temperature: Nov 1994 – Aug 1995")
plt.tight_layout()
plt.show()
```

## Air temperature over entire 1994 to 1998 period

```python
# Air temperature over entire 1994 to 1998 period, plotted cronologically
files = {
    1994: f"{BASE_DIR}/data/data_processed/climate_data_94.csv",
    1995: f"{BASE_DIR}/data/data_processed/climate_data_95.csv",
    1996: f"{BASE_DIR}/data/data_processed/climate_data_96.csv",
    1997: f"{BASE_DIR}/data/data_processed/climate_data_97.csv",
    1998: f"{BASE_DIR}/data/data_processed/climate_data_98.csv"
}

fig, ax = plt.subplots(figsize=(12, 6))

for year, file in files.items():
    df = pd.read_csv(file)
    
    # convert day_of_year to datetime
    df["Date"] = pd.to_datetime(f"{year}-01-01") + pd.to_timedelta(df["day_of_year"] - 1, unit="D")
    
    df = df.set_index("Date")
    
    # plot each year
    ax.plot(df.index, df["air_temp_average"], label=str(year), alpha=0.85)

# format x-axis as months only
ax.xaxis.set_major_formatter(DateFormatter('%b'))
plt.xticks(rotation=0)

plt.xlabel("Month")
plt.ylabel("Average Air Temperature (°C)")
plt.title("Daily Air Temperature Comparison (1994–1998)")
plt.legend(title="Year")
plt.tight_layout()
plt.show()
```

```python
# Air temperature over entire 1994 to 1998 period, plotted superimposed
files = {
    1994: f"{BASE_DIR}/data/data_processed/climate_data_94.csv",
    1995: f"{BASE_DIR}/data/data_processed/climate_data_95.csv",
    1996: f"{BASE_DIR}/data/data_processed/climate_data_96.csv",
    1997: f"{BASE_DIR}/data/data_processed/climate_data_97.csv",
    1998: f"{BASE_DIR}/data/data_processed/climate_data_98.csv"
}

fig, ax = plt.subplots(figsize=(12, 10))

reference_year = 2000  # dummy year used only for alignment

for year, file in files.items():
    df = pd.read_csv(file)

    # Map day_of_year into the same reference year
    df["Date"] = (
        pd.to_datetime(f"{reference_year}-01-01") +
        pd.to_timedelta(df["day_of_year"] - 1, unit="D")
    )

    ax.plot(df["Date"], df["air_temp_average"], label=str(year), alpha=0.85)

# Force axis to exactly one Jan–Dec span
ax.set_xlim(pd.Timestamp(f"{reference_year}-01-01"),
            pd.Timestamp(f"{reference_year}-12-31"))

# Show only months (no year)
ax.xaxis.set_major_locator(MonthLocator())
ax.xaxis.set_major_formatter(DateFormatter("%b"))  # Jan, Feb, Mar...

plt.xlabel("Month")
plt.ylabel("Average Air Temperature (°C)")
plt.title("Seasonal Air Temperature Comparison (1994–1998)")
plt.legend(title="Year")
plt.tight_layout()
plt.show()
```

## Temperature response curve for elongation scaling

```python
tmin = 2.0
topt = 16.3
tmax = 25.0

temps = np.linspace(tmin - 2.0, tmax + 2.0, 500)
valid = (temps >= tmin) & (temps <= tmax)

if not (tmin < topt < tmax):
    raise ValueError("topt must be strictly between tmin and tmax")

# Same piecewise temperature response as compute_temperature_scaling()
# and Clausnitzer & Hopmans (1994), as used in the paper.
mid = 0.5 * (tmin + tmax)
if topt < mid:
    T = (temps - tmin) / (tmax - tmin)
    Topt = (topt - tmin) / (tmax - tmin)
else:
    T = (temps - tmax) / (tmin - tmax)
    Topt = (topt - tmax) / (tmin - tmax)

# compute sigma such that Topt ** sigma = 0.5
sigma = np.log(0.5) / np.log(Topt)

# final scale: sin(pi * T**sigma), with zero response outside the valid domain
scales = np.zeros_like(temps)
scales[valid] = np.sin(np.pi * (T[valid] ** sigma))
scales = np.clip(scales, 0.0, 1.0)

plt.figure(figsize=(8,5))
plt.plot(temps, scales, lw=2)
plt.axvline(tmin, color='k', ls='--', alpha=0.5)
plt.axvline(topt, color='k', ls=':', alpha=0.7)
plt.axvline(tmax, color='k', ls='--', alpha=0.5)
plt.ylim(-0.05, 1.05)
plt.xlim(temps.min(), temps.max())
plt.xlabel("Soil temperature (°C)")
plt.ylabel("Scaling factor (relative root elongation)")
plt.title("Temperature response curve (paper/compute_temperature_scaling)")
plt.grid(alpha=0.3)
plt.show()
```
